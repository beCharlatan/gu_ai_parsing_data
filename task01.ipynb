## Практическое задание №1

1. Посмотреть документацию к API GitHub, разобраться как вывести список репозиториев для конкретного пользователя, сохранить JSON-вывод в файле *.json.

```python
import requests

username = 'beCharlatan'

with open('repos.json', 'wb') as f:
  response = requests.get(f'https://api.github.com/users/{username}/repos')
  print(response.text)
  f.write(response.content)
```

2. Изучить [список открытых API](https://www.programmableweb.com/category/all/apis). Найти среди них любое, требующее авторизацию (любого типа). Выполнить запросы к нему, пройдя авторизацию. Ответ сервера записать в файл.

```python
import json

# input your github personal access token
token = '<personal access token>'

def createRepo(name, desc):
  repoDict = {'name': name, 'description': desc}
  response = requests.post('https://api.github.com/user/repos', auth=(username, token), data=json.dumps(repoDict))
  return response


with open('createRepoResponse.json', 'wb') as f:
  response = createRepo('task01-test-repo', 'Creating repo through github api using requests')
  print(response)
  f.write(response.content)
```
